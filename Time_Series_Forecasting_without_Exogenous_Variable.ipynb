We will be performing Forecasting on the amazon stock price. We will be droppin an Exogenous Variales along they way.

Dataset: https://www.kaggle.com/datasets/szrlee/stock-time-series-20050101-to-20171231


```python
import pandas as pd
```

```python
df = pd.read_csv('/content/drive/MyDrive/AMZN_2006-01-01_to_2018-01-01.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
#converting Date column from object to datetime
df['Date']=pd.to_datetime(df['Date'])
#we'll set this date as the index
#lets make the date column as the index
df.set_index('Date',inplace=True)
```

```python
df.info()
```

```python
#as the name is same for the entire data we can drop in
#we want to predict the closing price we will delete all exogenous variables
df.drop(['Name','Open','Low','High','Volume'],axis=1,inplace=True)
```

```python

df.shape
```

```python
#conver timeseries in to monthly ferquency
df = df.asfreq('M')
```

```python
df.head()
```

```python
#initialize a time series experiment
exp=TSForecastingExperiment()
```

```python
#set up the experiment, fh 12 means we want to forecast the data for next 12 months
exp.setup(data=df, target='Close' ,fh=12, session_id=152,numeric_imputation_target="nearest")
```

```python
#plotting the time series
exp.plot_model(plot='ts')
```

```python
#generating all graphs needed for understanding the time series
exp.plot_model(plot="diagnostics", fig_kwargs={"height": 800, "width": 1000})
```

```python
#let us check the staistics of this time series
exp.check_stats(test="summary")
```

```python
#chec the available models we can train on this data
exp.models()
```

```python
#let us create an arima model and a xgboost_csd_dt model
arima_model = exp.create_model('arima')
```

```python
#let us output a sample forecast
y_pred = exp.predict_model(arima_model)
```

```python
y_pred
```

```python
#plot the sample forecast
exp.plot_model(arima_model)
```

```python
#performing tunning on the arima model
tuned_arima_model = exp.tune_model(arima_model)
```

```python
#compare the tuned and untuned mode
exp.plot_model([arima_model, tuned_arima_model], data_kwargs={"labels": ["Baseline", "Tuned"]})
```

```python
#using automl for finding a list of values, compare the best 3 models
baseline_models = exp.compare_models(n_select=3)
```

```python
#save the metrics of all the models trained so far in the experiement
mertics_comparission = exp.pull()
```

```python
#tune the best models
best_tuned_models = [exp.tune_model(model) for model in baseline_models]
```

```python
#assign the top 3 model weights based on their MAE
top_model_metrics = mertics_comparission.iloc[0:3]['MAE']
display(top_model_metrics)
top_model_weights = 1 - top_model_metrics/top_model_metrics.sum() #weight of each model's prediction depends on its standalone MAE, lower the MAE higher its weight in the ensemble model
```

```python
display(top_model_weights)
```

```python
best_tuned_models
```

```python
#create a blended model
blended_model = exp.blend_models(best_tuned_models, method='mean', weights=top_model_weights.values.tolist())
```

```python
y_pred = exp.predict_model(blended_model)
```

```python
y_pred.shape
```

```python
exp.plot_model(blended_model)#ppredicting for the last 12 months
```

```python
#finalizing the model
final_model = exp.finalize_model(blended_model)# we'll use this to predict 12 monthss in future
```

```python
exp.plot_model(final_model)
```

```python
#save the model
exp.save_model(final_model, "/content/drive/MyDrive/ensemble_forecasting")
```
