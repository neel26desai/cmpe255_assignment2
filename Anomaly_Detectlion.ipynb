We will be performing anomally detection on NYC Taxi Traffic data .
Kaggle link: https://www.kaggle.com/datasets/julienjta/nyc-taxi-traffic
The Traffic is bined to 30 minutes interval and expected anomalies occur during the NYC marathon, Thanksgiving, Christmas, New Years day, and a snow storm

```python
#loading the data set
import pandas as pd
df= pd.read_csv('/content/drive/MyDrive/dataset.csv', usecols=['timestamp','value'])
```

```python
#cast timestap as time and not string
df['timestamp']=pd.to_datetime(df['timestamp'])#need to do a manual conversion or else the set up takes too long
```

```python
df.info()
```

```python
#import AnomalyExperiment and initialize its class
from pycaret.anomaly import AnomalyExperiment
exp = AnomalyExperiment()
```

```python
#setup the experiment
exp.setup(df,session_id=123,log_experiment = False)
```

```python
#lets check the list of models we have, that can be used
exp.models()
```

```python
#let us a train a iforest,svm and cluster based outlier
iforest_model = exp.create_model('iforest')
svm_model= exp.create_model('svm')
cluster_model = exp.create_model('cluster')
```
